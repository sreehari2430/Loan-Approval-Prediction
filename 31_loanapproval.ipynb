```python
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
data=pd.read_csv('/content/drive/MyDrive/Datasets/LoanApprovalPrediction.csv')
data
```

```python
X=data.iloc[:,1:-1]
y=data['Loan_Status']
```

```python
X.dtypes
```

```python
cat_features = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area']
num_features = ['LoanAmount', 'Loan_Amount_Term', 'Credit_History']
cat_transform=Pipeline(steps=[
    ('si',SimpleImputer(strategy='most_frequent')),
    ('encoder',OneHotEncoder())
])
num_transform=Pipeline(steps=[
    ('si',SimpleImputer(strategy='mean')),
    ('scaler',MinMaxScaler()),
])
```

```python
preprocessor = ColumnTransformer( transformers=[
    (('cat', cat_transform, cat_features)),
    ('num', num_transform, num_features)
])
```

```python
pipeline=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('rfc',RandomForestClassifier(random_state=1,max_depth=3))
])
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=5)
```

```python
pipeline.fit(X_train,y_train)
```

```python
y_pred=pipeline.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print("Training Score: ", pipeline.score(X_train, y_train))
```

```python
print("Testing Score: ", pipeline.score(X_test, y_test))
```
